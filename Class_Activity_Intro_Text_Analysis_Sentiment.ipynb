## We are creating a function we will be repeatedly calling to print the Sentiment Score of any given sentence in a structured format

```python
analyser = SentimentIntensityAnalyzer()

def print_sentiment_scores(sentence):
    snt = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(snt)))
```

```python
print_sentiment_scores("The movie is not good")
```

```python
print_sentiment_scores("I was asked to meet in the evening - does he realise it's my birthday?")

# What happens if you add a smiley?
print_sentiment_scores("I was asked to meet in the evening - does he realise it's my birthday? :(")

#Add an acronym "smh" - "shaking my head"
print_sentiment_scores("I was asked to meet in the evening - does he realise it's birthday? smh :(")

#context
print_sentiment_scores("The food is good.")

#context
print_sentiment_scores("The food is bad")

#context
print_sentiment_scores("The food is BAD!")

# if one word is capitalized
print_sentiment_scores("The food is GOOD.")

#An exclamation added
print_sentiment_scores("The food is GOOD!")
```

## Importing the Natural Language Toolkit - Naive Bayes Classifier

```python
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
```

```python
def word_features(words):
    return dict([(word, True) for word in words])
```

### Creating Positive, Negative, and Neutral word subsets

```python
positive_vocabulary = [ 'awesome', 'outstanding', 'fantastic', 'terrific', 'good', 'nice', 'great', ':)', 'liked' ]
negative_vocabulary = [ 'bad', 'terrible','useless', 'hate', ':(' ]
neutral_vocabulary = [ 'movie','the','sound','was','is','actors','did','know','words','not' ]
```

### Feature Extraction - Essentially, for each word, we are keeping track of a positive feature set {& negative and neutral ones}

```python
positive_features = [(word_features(pos), 'pos') for pos in positive_vocabulary]
negative_features = [(word_features(neg), 'neg') for neg in negative_vocabulary]
neutral_features = [(word_features(neu), 'neu') for neu in neutral_vocabulary]
```

### Train the classifier with the word sets we created

```python
train_set = negative_features + positive_features + neutral_features

classifier = NaiveBayesClassifier.train(train_set)
```

```python
# Predict
neg = 0
pos = 0

# You can try forming sentences here for the program to generate positivity and negativity scores
sentence = "terrible movie outstanding"
sentence = sentence.lower()
words = sentence.split(' ')
# Python understands that sentence is a string and tokenizes it into words

for word in words:
    classResult = classifier.classify(word_features(word))
    print(classResult)
    if classResult == 'neg':
        neg = neg + 1
    if classResult == 'pos':
        pos = pos + 1
```

### Print the proportion of positive and negative words out of total - which is basically the sentiment score

```python
print('Positive: ' + str(float(pos)/len(words)))
print('Negative: ' + str(float(neg)/len(words)))
```
